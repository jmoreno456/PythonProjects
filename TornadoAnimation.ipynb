Import Libraries

```python
import pandas as pd
import plotly.express as px
```

load the dataset

```python
data = pd.read_csv("TornadoData1.csv")
print(data.head())
print(data.describe())
```

filter and group the data

```python
tornado_data = (
    data[data["mag"] >= 0] # filter rows which contain negative values
    .groupby(["yr", "st"]) # group by year and state
    .size() # count the number of tornadoes
    .reset_index(name = "total_tornadoes") # rename count column
)
```

create animated bubble chart

```python
fig = px.scatter (
    tornado_data, 
    x = "st", 
    y = "total_tornadoes", 
    color = "st", 
    animation_frame = "yr", 
    size_max = 40, 
    title = "Total Tornadoes per State Over Time", 
    labels = {"st": "State", "total_tornadoes": "Total Tornadoes"}
)

fig.update_layout (
    xaxis = dict(title = "State"), 
    yaxis = dict(title = "Total Tornadoes"), 
    showlegend = False, 
    template = "simple_white"
)
```
